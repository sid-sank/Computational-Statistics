# Functions and control operations


```python
def has_passed(int): # Local and global var names allow this
  if int > 80: # This int var is local you cannot call it outside this function
    return True
  else: return False

Pass_TF = [has_passed(i) for i in scores] # Called list comprehension - new way

for i in scores: # Old python way
  print(f"student #{scores.index(i) + 1}: {has_passed(i)}")

def TF_to_SU(bool):
  if bool: return "Satisfactory"
  else: return "Unsatisfactory"
SU = [TF_to_SU(i) for i in Pass_TF]
print(SU)
```

# Match statements

```python
def get_grades(int):
  match int:
    case int if int < 40:
      return "U"
    case int if 40 <= int < 50:
      return "E"
    case int if 50 <= int < 60:
      return "D"
    case int if 60 <= int < 70:
      return "C"
    case int if 70 <= int < 80:
      return "B"
    case int if 80 <= int < 90:
      return "A-"
    case int if 90 <= int < 95:
      return "A"
    case int if 95 <= int :
      return "O"
grades = [ get_grades(i) for i in scores]
print(grades)
```

# Array operations


```python
arr1 = np.array([1, 2, 3, 4, 5])
arr2 = np.array([6, 7, 8, 9, 10])
print(f"First array is {arr1}, second array is {arr2}")
print(f"They are both of type {type(arr1)}, not a regular list")
```

```python
sum_arr = arr1 + arr2 # Elementwise addition
sub_arr = arr1 - arr2 # Elementwise subtraction
mul_arr = arr1 * arr2 # Elementwise multiplication
div_arr = arr1 / arr2 # Elementwise division

print(sum_arr, sub_arr, mul_arr, div_arr)
```
