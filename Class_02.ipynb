# Functions and control operations


```python
def has_passed(int): # Local and global var names allow this
  if int > 80: # This int var is local you cannot call it outside this function
    return True
  else: return False

Pass_TF = [has_passed(i) for i in scores] # Called list comprehension - new way

for i in scores: # Old python way
  print(f"student #{scores.index(i) + 1}: {has_passed(i)}")

def TF_to_SU(bool):
  if bool: return "Satisfactory"
  else: return "Unsatisfactory"
SU = [TF_to_SU(i) for i in Pass_TF]
print(SU)
```

# Match statements

```python
def get_grades(int):
  match int:
    case int if int < 40:
      return "U"
    case int if 40 <= int < 50:
      return "E"
    case int if 50 <= int < 60:
      return "D"
    case int if 60 <= int < 70:
      return "C"
    case int if 70 <= int < 80:
      return "B"
    case int if 80 <= int < 90:
      return "A-"
    case int if 90 <= int < 95:
      return "A"
    case int if 95 <= int :
      return "O"
grades = [ get_grades(i) for i in scores]
print(grades)
```

```python
prices = np.array([230, 450, 680, 970, 330, 850, 200, 130, 50])
print(prices[1])    # Prints 2nd element
print(prices[3:6])  # Prints indexs 3, 4, 5. Start is inclusive, end is not
print(prices[3:])   # Prints from the fourth element onwards (index = 3)
print(prices[:6])   # Prints until the 6th element (index = 5)
print(prices[-1])   # Prints last element
print(prices[:-2])  # Prints until 2nd last element
print(prices[-3:])  # Prints last 3 elements
print(prices[::2])  # Prints every secod element starting from index = 0
print(prices[::-1]) # Prints list in reverse
```
